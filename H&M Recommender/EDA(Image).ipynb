# Pre-processing images

```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#How to preprocess the images
def preprocess_image(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    image = cv2.equalizeHist(image)
    image = image/255
    return image
```

```python
#Bringing the images into the notebook
def load_images_from_folder(folder):
    images = []
    labels = []
    for filename in os.listdir(folder):
        label = filename.split('.')[0]
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            img = preprocess_image(img)
            images.append(img)
            labels.append(label)
    return images, labels
```

```python
image_dir = "data/images"
```

```python
def getImage_path(path):
    image_path = []
    for filename in os.listdir(path):
        image_path.append(os.path.join(path,filename))
    return image_path
```

```python
#Write a function to read the images and labels
def load_images_and_labels(image_folder):
    images = []
    labels = []
    for folder in image_folder:
        image, label = load_images_from_folder(folder)
        images.append(image)
        labels.append(label)
    return images, labels
```
